```python
import matplotlib.pyplot as plt

months = [1,2,3,4,5]
sales = [100,212,123,500,400]
plt.plot(months , sales, color='blue', linestyle='--', linewidth=2 , marker='o',label='sales data')
plt.title("monthly sales data")
plt.ylabel('sales')
plt.xlabel('months')
plt.xlim(1,4)
plt.ylim(1,600)
plt.xticks([1,2,3,4,5],['M1', 'M2', 'M3', 'M4','M5'])
plt.legend(loc = 'upper left', fontsize=10)
plt.grid(color='green',linestyle=':',linewidth=1)
plt.show()
```

```python
import matplotlib.pyplot as plt
product = ['A','B','C','D']
sales = [1000,2000,300,600]
plt.barh(product,sales, color='pink',label='sales of 2025')
plt.xlabel('product')
plt.ylabel('sales')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
region = ['North','South','East','West']
revene = [1000,2000,3000,2500]
plt.pie(revene, labels=region,autopct='%1.1f%%', colors=['gold','skyblue','green','orange'])
plt.show()
```

```python
import matplotlib.pyplot as plt
scores = [30,40,50,60,70,80,90,10,20,34,32,77,74,34,73,11,68,74,33]
plt.hist(scores, bins=10, color='purple', edgecolor='white')
plt.xlabel('scores')
plt.show()
```

```python
import matplotlib.pyplot as plt
hour_study = [1,2,3,4,5,6,7]
marks = [50,60,70,80,90,95,100]
plt.scatter(hour_study,marks , color='red', label='Student marks data',marker='^')
plt.xlabel('hours of study')
plt.ylabel('marks')
plt.legend()
plt.title('students data')
plt.grid(True)
plt.show()
```
